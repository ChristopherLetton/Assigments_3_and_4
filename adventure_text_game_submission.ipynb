```python
# Project Brief
# Remember Adventure? Well, we’re going to build a more basic version of that. A complete text
# game, the program will let users move through rooms based on user input and get descriptions
# of each room. To create this, you’ll need to establish the direction in which the user can move, a
# way to track how far the user has moved (and therefore which room he/she is in), and to print
# out a description. You’ll also need to set limits for how far the user can move. In other words,
# create “walls” around the rooms that tell the user, “You can’t move further in this direction.”

# Define rooms with their descriptions
rooms = {"SPAWN" : [("You are in the 1st room. You can go NORTH, SOUTH, EAST or WEST."), ("")],
         "NORTH" : [("You have moved NORTH. It's too dark; you can't see anything."), ("")],
         "EAST" : [("You have moved EAST. There is a scary bunny."), ("")] ,
         "WEST" : [("You have moved WEST. There are no other doors in this room"), ("")],
         "SOUTH" : [("You have espaced! Did you see the bunny???"),("")]}

# Set to starting room
current_room = "SPAWN"

# Narrative before initiating main loop
print("You awake in a strange room with no recollection of how you got here. QUICK ESCAPE!")
print(rooms["SPAWN"][0])

# Main game loop
while True:
    # User inputs desired direction
    move = input("Which direction do you want to go? NORTH, SOUTH, EAST or WEST").upper()
    # Checks if input is in a valid direction
    if move in ["NORTH", "SOUTH", "EAST", "WEST"]:
            # Checks the current room and then updates accordingly 
            if current_room == "SPAWN" and move == "NORTH":
                current_room = "NORTH"
                print(rooms[current_room][0])
                continue
            elif current_room == "NORTH" and move == "SOUTH":
                current_room = "SPAWN"
                print(rooms[current_room][0])
            elif current_room == "SPAWN" and move == "SOUTH":
                current_room = "SOUTH"
                print(rooms[current_room][0])
                break
            elif current_room == "SPAWN" and move == "EAST":
                current_room = "EAST"
                print(rooms[current_room][0])
                continue
            elif current_room == "EAST" and move == "WEST":
                current_room = "SPAWN"
            elif current_room == "SPAWN" and move == "WEST":
                current_room = "WEST"
                print(rooms[current_room][0])
                continue
            elif current_room == "WEST" and move == "EAST":
                current_room = "SPAWN"
                print(rooms[current_room][0])
            else:
                print("The exit isn't that way! Go back!") 
                continue
    else:
        print("Do you not want to escape??? Try entering north,south, east or west!")
        continue
```
